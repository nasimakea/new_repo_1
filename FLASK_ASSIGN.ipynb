Q3. What is App routing in Flask? Why do we use app routes?

ANS:-


App routing in Flask refers to the mechanism of mapping URLs (or routes) to specific functions within a web application. These routes are defined using the @app.route() decorator in Flask. They help direct incoming requests to the appropriate view functions, allowing the application to respond accordingly. App routes are essential as they enable developers to create a logical structure within their Flask application, defining what action should be taken when a specific URL is accessed. This organization simplifies the management of different functionalities and enhances the overall readability and maintainability of the codebase.
